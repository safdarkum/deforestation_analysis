```python
#Descriptive Statistics (Frequency )

#Frequency of Gender
# Importing Pandas
import pandas as pd
#Loading the CSV file

df = pd.read_csv("data.csv") 

# Mapping codes to the variables
code_map = {
    0: 'Female', 
    1: 'Male',
}

df['qty_range'] = df['Gender'].map(code_map)

# Frequency 
frequency = df['qty_range'].value_counts().sort_index()
print(frequency)
```

```python
#Frequency of Residence (Urban or Rural)
#Mapping codes to the variables
code_map = {
    0: 'Rural', 
    1: 'Urban',
}

df['qty_range'] = df['Residence_area'].map(code_map)

# Frequency 
frequency = df['qty_range'].value_counts().sort_index()
print(frequency)
```

```python
#Frequency of Education
#Mapping codes to the variables
code_map = {
    1: 'Illiterate', 
    2: 'Metric',
    3: 'Under Graduate',
    4: 'Post Graduate'  
}

df['qty_range'] = df['Education_level'].map(code_map)

# Frequency 
frequency = df['qty_range'].value_counts().sort_index()
print(frequency)
```

```python
#Frequency of Land ownership
#Mapping codes to the variables
code_map = {
    0: 'No Land', 
    1: 'Land', 
}

df['qty_range'] = df['Land Status'].map(code_map)

# Frequency 
frequency = df['qty_range'].value_counts().sort_index()
print(frequency)
```

```python
#Frequency of land use type
# Defining the binary columns
land_use_columns = ['Land_agriculture', 'Land_forest', 'Land_range', 'Land_barren']

# Initializing an empty list for results
rows = []

# Loop through each column and count 0s and 1s
for col in land_use_columns:
    count_0 = (df[col] == 0).sum()
    count_1 = (df[col] == 1).sum()
    rows.append([col, count_0, count_1])

# Creating  DataFrame from results
frequency_df = pd.DataFrame(rows, columns=['Land_Type', 'No (0)', 'Yes (1)'])

#'Land_Type' as index 
frequency_df.set_index('Land_Type', inplace=True)

# Print frequenc
print(frequency_df)
```

```python
#Frequency of Income type of the respondents 
#Defining the binary columns
land_use_columns = ['Income_agriculture', 'Income_forest', 'Income_livestock', 'Income_govt', 'Income_business']

# Initializing an empty list for results
rows = []

# Loop through each column and count 0s and 1s
for col in land_use_columns:
    count_0 = (df[col] == 0).sum()
    count_1 = (df[col] == 1).sum()
    rows.append([col, count_0, count_1])

# Creating DataFrame from results
frequency_df = pd.DataFrame(rows, columns=['Income_Type', 'No (0)', 'Yes (1)'])

#'Land_Type' as index 
frequency_df.set_index('Income_Type', inplace=True)

# Print frequency
print(frequency_df)
```

```python
#Frequency of Fuelwood dependance
#Mapping codes to variables
code_map = {
    0: 'No fuel wood', 
    1: 'Fuel Wood uses', 
}

df['qty_range'] = df['Fuelwood_Dependance'].map(code_map)

# Frequency
frequency = df['qty_range'].value_counts().sort_index()

# Display frequency
print(frequency)
```

```python
#Frequency of daily Fuelwood collection
#Mapping codes to variables
code_map = {
    1: '1 to 25kg', 
    2: '25 to 50 kg', 
    3: '50 to 80 kg',
    4: 'More than 80 kg'
}

df['qty_range'] = df['Fuelwood_qty'].map(code_map)

# Frequency 
frequency = df['qty_range'].value_counts().sort_index()

# Display frequency
print(frequency)
```

```python
#Frequency of Fuelwood usage by the respondents 
#Defining the binary columns
land_use_columns = ['Fuel_use_cooking',	'fuel_use_warming',	'fuel_use_lighting', 'fuel_use_ironing', 'fuel_use_bonfire']

# Initializing an empty list for results
rows = []

# Loop through each column and count 0s and 1s
for col in land_use_columns:
    count_0 = (df[col] == 0).sum()
    count_1 = (df[col] == 1).sum()
    rows.append([col, count_0, count_1])

# Creating a DataFrame from results
frequency_df = pd.DataFrame(rows, columns=['Fuelwood_usage', 'No (0)', 'Yes (1)'])

# 'Fuelwood usage' as index 
frequency_df.set_index('Fuelwood_usage', inplace=True)

# Print frequency
print(frequency_df)
```

```python
#Frequency of Fuelwood usage by the respondents
#Mapping codes to variables
code_map = {
    1: 'Always', 
    2: 'Occasionally', 
    3: 'Not at all',
    4: 'Once a while'
}

df['qty_range'] = df['Freq_fuelwood'].map(code_map)

# Frequency 
frequency = df['qty_range'].value_counts().sort_index()

# Display frequency
print(frequency)
```

```python
#Frequency of Political interference to Forest Department
#Mapping codes to variables
code_map = {
    0: 'Political', 
    1: 'No Political', 
}

df['qty_range'] = df['Political_interference'].map(code_map)

# Frequency count
frequency = df['qty_range'].value_counts().sort_index()

# Display frequency
print(frequency)
```
